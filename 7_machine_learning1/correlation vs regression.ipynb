```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
from scipy import *
```

```python
x = np.array([1, 3, 6, 10, 15])
y = np.array([89, 85, 79, 73, 64])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
corr, _ = scipy.stats.pearsonr(x, y)
print(corr)
```

## Relation between regression and correlation

### Slopes dont change correlation significantly

```python
x = list(range(1, 50, 1))
y = x
```

```python
plt.scatter(x,y)
plt.show()
```

```python
corr, _ = scipy.stats.pearsonr(x, y)
print(corr)
```

```python
x = list(range(1, 50, 1))
y = [i/2 for i in x]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
corr, _ = scipy.stats.pearsonr(x, y)
print(corr)
```

```python
x = list(range(1, 10, 1))
y = [4]*len(x)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
corr, _ = scipy.stats.pearsonr(x, y)
print(corr)
```

```python
dff = pd.read_csv('odi-batting.csv')
```

```python
dff.head()
```

```python
dff['centuries'] = (dff.Runs >= 100)
```

```python
dff['centuries'] = dff['centuries'].astype('int')
```

```python
dff['centuries']
```

```python
dff.pivot_table(index='Player', values='centuries', aggfunc=np.sum).sort_values(ascending=False, by='centuries')
```

```python
dff1 = dff[dff['centuries']>0]
```

```python
dff1['sr'] = dff1.Runs.div(dff1.Balls)*100
```

```python
dff1.sort_values(ascending=False, by='sr')
```

```python
dff2 = dff1[dff1['Country']=='India'].copy()
```

```python
dff2['year'] = dff2['MatchDate'].apply(lambda x: x.split('-')[-1])
```

```python
dff2
```

```python
dff2.pivot_table(index='year', values='centuries', aggfunc=np.sum).sort_values(ascending=False, by='centuries')
```

## date, day, month, year, hour, and minutes

```python
df_u = pd.read_csv('grades.csv')
```

```python
df_u
```

```python
df_u['format']=df_u.submission.apply(lambda x: x.split('.')[-1])
```

```python
df_u.format.isin(['zip']).sum()/df_u.shape[0]
```

```python
df_u['day'] = df_u.submit_time.apply(lambda x: x.split('-')[0].split('/')[1])
```

```python
df_u['month'] = df_u.submit_time.apply(lambda x: x.split('-')[0].split('/')[0])
```

```python
df_u['year'] = df_u.submit_time.apply(lambda x: x.split('-')[0].split('/')[2])
```

```python
df_u['hour'] = df_u.submit_time.apply(lambda x: x.split('-')[1].split(':')[0])
```

```python
df_u['min'] = df_u.submit_time.apply(lambda x: x.split('-')[1].split(':')[1])
```

```python
df_u
```

```python
df_u_sub  = df_u[(df_u['year']=='17')&(df_u['month']=='01')]
```

```python
df_u_sub['day'] = df_u_sub['day'].astype('int')
```

```python
df_u_sub['hour'] = df_u_sub['hour'].astype('int')
```

```python
df_u_sub['min'] = df_u_sub['min'].astype('int')
```
